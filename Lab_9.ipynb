# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):

    distances = {}
    visited = []

    for node in graph:
        distances[node] = float('inf')

    distances[start] = 0

    while len(visited) < len(graph):

        min_node = None
        min_distance = float('inf')

        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        visited.append(min_node)

        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight

            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance

    return distances
```

```python
graph = {
'A': {'B': 4, 'C': 2},
'B': {'A': 4, 'C': 5, 'D': 10},
'C': {'A': 2, 'B': 5, 'D': 3},
'D': {'B': 10, 'C': 3}
}

result = dijkstra(graph, 'A')

for node in result:
    print("Distance from A to", node, "=", result[node])
```

```python
def dijkstra(graph, start):
    distances = {node: float('inf') for node in graph}
    previous_nodes = {node: None for node in graph}
    visited = []

    distances[start] = 0

    while len(visited) < len(graph):
        min_node = None
        min_distance = float('inf')

        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        if min_node is None:
            break

        visited.append(min_node)

        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight

            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance
                previous_nodes[neighbor] = min_node

    return distances, previous_nodes

def reconstruct_path(previous_nodes, start, target):
    """Backtracks through the previous_nodes dictionary to build the path."""
    path = []
    current_node = target

    while current_node is not None:
        path.append(current_node)
        current_node = previous_nodes[current_node]

    path.reverse()

    if path and path[0] == start:
        return " -> ".join(path)
    return "No path exists"

def input_custom_graph():
    """Allows the user to build a graph dictionary via console input."""
    graph = {}
    print("\n--- Create Your Custom Graph ---")
    print("First, define all the node names.")
    nodes = input("Enter all node names separated by spaces (e.g., A B C): ").split()

    for node in nodes:
        graph[node] = {}

    print("\nNow, define the edges. Type 'done' when finished.")
    while True:
        edge_input = input("Enter Edge (format: Node1 Node2 Weight) or 'done': ").strip()
        if edge_input.lower() == 'done':
            break

        try:
            u, v, w = edge_input.split()
            weight = int(w)

            if u not in graph or v not in graph:
                print(f"Error: Both nodes must be in your defined list: {nodes}")
                continue

            graph[u][v] = weight
            graph[v][u] = weight
            print(f"Added edge: {u} <-> {v} with weight {weight}")
        except ValueError:
            print("Invalid format. Please use: Node1 Node2 Weight (e.g., A B 4)")

    return graph

def main():
    default_graph = {
        'A': {'B': 4, 'C': 2},
        'B': {'A': 4, 'C': 5, 'D': 10, 'E': 3},
        'C': {'A': 2, 'B': 5, 'D': 3, 'F': 8},
        'D': {'B': 10, 'C': 3, 'E': 4, 'F': 2},
        'E': {'B': 3, 'D': 4, 'F': 6},
        'F': {'C': 8, 'D': 2, 'E': 6}
    }

    print("--- Dijkstra's Shortest Path Algorithm ---")
    choice = input("Do you want to use the default 6-node graph? (y/n): ").strip().lower()

    if choice == 'n':
        graph = input_custom_graph()
    else:
        graph = default_graph

    if not graph:
        print("Graph is empty. Exiting.")
        return

    while True:
        print(f"\nAvailable nodes: {', '.join(graph.keys())}")
        start_node = input("Enter the START node (or 'quit' to exit): ").strip()

        if start_node.lower() == 'quit':
            break
        if start_node not in graph:
            print("Invalid start node.")
            continue

        target_node = input("Enter the TARGET node: ").strip()
        if target_node not in graph:
            print("Invalid target node.")
            continue

        distances, previous_nodes = dijkstra(graph, start_node)

        print("\n" + "="*40)
        if distances[target_node] == float('inf'):
            print(f"❌ No path exists from {start_node} to {target_node}.")
        else:
            path_string = reconstruct_path(previous_nodes, start_node, target_node)
            print(f"✅ Shortest Path from {start_node} to {target_node}: {path_string}")
            print(f"📏 Total Distance: {distances[target_node]}")
        print("="*40)

if __name__ == "__main__":
    main()
```
